```python
from langchain_core.tools import tool
```

```python
# Step 1
def multiply(a,b):
    """Multiply two nos"""
    return a*b
```

```python
# Step 2
def multiply(a:int,b:int)->int:
    """Multiply two nos"""
    return a*b
```

```python
# Step 3

@tool
def multiply(a:int,b:int)->int:
    """Multiply two nos"""
    return a*b
```

```python
result = multiply.invoke({'a':3,'b':4})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```
