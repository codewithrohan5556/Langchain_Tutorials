```python
from langchain_community.vectorstores import FAISS
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_core.documents import Document
from langchain_openai import ChatOpenAI
from langchain_classic.retrievers.multi_query import MultiQueryRetriever
from dotenv import load_dotenv
load_dotenv()
```

```python
docs = [
    Document(page_content="",metadata={'source':'H1'}),
    Document(page_content="",metadata={'source':'H2'}),
    Document(page_content="",metadata={'source':'H3'}),
    Document(page_content="",metadata={'source':'H4'}),
    Document(page_content="",metadata={'source':'H5'}),
    Document(page_content="",metadata={'source':'I1'}),
    Document(page_content="",metadata={'source':'I2'}),
    Document(page_content="",metadata={'source':'I3'}),
    Document(page_content="",metadata={'source':'I4'}),
    Document(page_content="",metadata={'source':'I5'})
]
```

```python
embedding_model = OpenAIEmbeddings()

vector_store = FAISS.from_documents(documents=docs,embedding=embedding_model)
```

```python
similarity_retriever = vector_store.as_retriever(search_type='similarity',search_kwargs={'k':5})
```

```python
multiquery_retriever = MultiQueryRetriever.from_llm(
    retriever=vector_store.as_retriever(search_kwargs={'k':5}),
    llm = ChatOpenAI(model='gpt-3.5-turbo')
)
```

```python
query = 'how to improve energy levels and maintain balance ?'
```

```python
similarity_results = similarity_retriever.invoke(query)
multiquery_results = multiquery_retriever.invoke(query)
```

```python
for i,doc in enumerate(similarity_results):
    print(f"\n---Result {i+1}----")
    print(doc.page_content)
    # normal results
```

```python
for i,doc in enumerate(multiquery_results):
    print(f"\n---Result {i+1}----")
    print(doc.page_content)
    # query related results no hallucination
```
