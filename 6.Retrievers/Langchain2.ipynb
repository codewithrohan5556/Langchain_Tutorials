```python
from abc import ABC,abstractmethod
class Runnable(ABC):
    @abstractmethod
    def invoke(input_data):
        pass
```

```python
import random

class NakliLLM(Runnable):
    def __init__(self):
        print('LLM Created')

    def invoke(self,prompt):
        response_list = [
            'Delhi is the capital of india',
            'IPL is stupid in india',
            'AI is going to take all jobs'
        ]
        return {'response':random.choice(response_list)}
    

    def predict(self,prompt):
        response_list = [
            'Delhi is the capital of india',
            'IPL is stupid in india',
            'AI is going to take all jobs'
        ]
        return {'response':random.choice(response_list),'Warning':'This method is been depreciated check once the offical documentation'}
    
```

```python
class NakliPromptTemplate(Runnable):

    def __init__(self,template,input_variables):
        self.template = template
        self.input_variables = input_variables
    def invoke(self,input_dict):
        return self.template.format(**input_dict)
    
    def format(self,input_dict):
        print('Warning:This method is been depreciated check once the offical documentation')
        return self.template.format(**input_dict)
```

```python
class NakliStrOutputParser(Runnable):
    def __init__(self):
        pass
    def invoke(self,input_data):
        return input_data['response']
```

```python
class RunnableConnector(Runnable):
    def __init__(self,runnable_list):
        self.runnable_list = runnable_list

    def invoke(self,input_data):
        for runnable in self.runnable_list:
            input_data = runnable.invoke(input_data)
        return input_data
```

```python
template = NakliPromptTemplate(
    template='write a {length} poem about {topic}',
    input_variables=['length','topic']
)
```

```python
llm = NakliLLM()
```

```python
parser = NakliStrOutputParser()
```

```python
chain = RunnableConnector([template,llm,parser])
```

```python
chain.invoke({'length':'short','topic':'india'})
```
